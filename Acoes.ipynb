```
import plotly.graph_objs as go

Close = go.Scatter(
x=df.Date,
y=df.Close,
name = "PETR4 Close",
line = dict(color = '#17BECF'),
opacity = 0.8)

data = [Close]

layout = dict(
title = "Faixa de Data Customizada",
title_x= 0.5,
xaxis = dict(
range = ['2020-01-01','2020-12-04'])
)

fig = dict(data=data, layout=layout)
py.iplot(fig, filename = "Manually Set Range")
```

```
import plotly.graph_objs as go

Close = go.Scatter(
x=df.Date,
y=df.Close,
name = "PETR4 High",
line = dict(color = '#17BECF'),
opacity = 0.8)

data = [Close]

layout = dict(
title="Srie com Rangeslider e Botoes",
title_x=0.5,
xaxis=dict(
rangeselector=dict(
buttons=list([
dict(count=1,
label='1m',
step='month',
stepmode='backward'),
dict(count=6,
label='6m',
step='month',
stepmode='backward'),
dict(step='all')
])
),
rangeslider=dict(
visible = True
),
type='date'
)
)

fig = dict(data=data, layout=layout)
py.iplot(fig)
```

```
trace = go.Candlestick(x=df['Date'],
open=df['Open'],
high=df['High'],
low=df['Low'],
close=df['Close'])

layout = go.Layout(
title ='Preos em Candlesticks',
title_x = 0.5,
xaxis = dict(
rangeslider = dict(
visible = False
)
)
)

data = [trace]

fig = go.Figure(data=data,layout=layout)
py.iplot(fig)
```

```
trace = go.Candlestick(x=df['Date'],
open=df['Open'],
high=df['High'],
low=df['Low'],
close=df['Close'],
increasing=dict(line=dict(color= '#17BECF')),
decreasing=dict(line=dict(color= '#7F7F7F'))
)

layout = go.Layout(
title ='Preos em Candlesticks',
title_x = 0.5,
xaxis = dict(
rangeslider = dict(
visible = False
)
)
)

data = [trace]

fig = go.Figure(data=data,layout=layout)
py.iplot(fig)
```

```
# Mdia simples de 3 dias
df['MM_3'] = df.Close.rolling(window=3).mean()

# Mdia simples de 9 dias
df['MM_9'] = df.Close.rolling(window=9).mean()

# Mdia simples de 17 dias
df['MM_17'] = df.Close.rolling(window=17).mean()

close = go.Scatter(
x=df.Date,
y=df.Close,
name = "PETR4 Close",
line = dict(color = '#330000'),
opacity = 0.8)

MM_3 = go.Scatter(
x=df.Date,
y=df['MM_3'],
name = "Mdia Mvel 3 Perodos",
line = dict(color = '#FF8000'),
opacity = 0.8)

MM_9 = go.Scatter(
x=df.Date,
y=df['MM_9'],
name = "Mdia Mvel 9 Perodos",
line = dict(color = '#B2FF66'),
opacity = 0.8)

MM_17 = go.Scatter(
x=df.Date,
y=df['MM_17'],
name = "Mdia Mvel 17 Perodos",
line = dict(color = '#FF00FF'),
opacity = 0.8)

data = [close, MM_3, MM_9, MM_17]

py.iplot(data)
```
